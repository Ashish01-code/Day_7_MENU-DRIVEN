```python
num=[]
while(True):
    print("CHOOSE AN OPEARTION")
    print("1. APPPEND")
    print("2. INSERT")
    print("3. REMOVE")
    print("4. POP")
    print("5. LENGTH")
    print("6. SUM")
    print("7. MAX AND MIN")
    print("8. SORT AND REVERSE SORT")
    print("9. COUNT")
    print("10. EXIT THE PROGRAM AND PRINT THE FINAL LIST")
    choice=int(input("ENTER OPERATION"))
    if choice==1:
        a=int(input("ENTER THE LIMIT TO ADD NUMBER"))
        print("ENTER NUMBER TO ADD")
        for i in range(a):
            b=int(input())
            num.append(b)
        print("THE LIST IS:",num)
    elif choice==2:
        index=int(input("ENTER THE INDEX "))
        value=int(input("ENTER THE NUMBER TO INSERT"))
        num.insert(index,value)
        print("THE LIST IS:",num)
    elif choice==3:
        a=int(input("ENTER THE NUMBER TO REMOVE"))
        if a in num:
            num.remove(a)
            print("THE LIST IS :",num)
        else:
            print("INVALID NUMBER OR NO SUCH NUMBER EXIST")
    elif choice==4:
        if len(num)>0:
            index=int(input("ENTER THE INDEX TO POP"))
            if 0<=index<len(num):
                num.pop(index)
                print("THE LIST IS:",num)
            else:
                print("INVALID INDEX")
        else:
            print("EMPTY LIST")
    elif choice==5:
        print("THE LENGTH:",len(num))
        print("THE LIST IS:",num)
    elif choice==6:
        print("THE SUM IS: ",sum(num))
        print("THE LIST IS:",num)
    elif choice==7:
        if len(num)>0:
            print("THE MAX:",max(num),"MIN IS:",min(num))
            print("THE LIST IS:",num)
        else:
            print("NO ELEMENTS SO NO MAX AND MIN")
    elif choice==8:
        num.sort()
        print("THE ASCENDING ORDER:",num)
        num.sort(reverse=True)
        print("THE DESCENDING ORDER:",num) 
     
    elif choice==9:
        value=int(input("ENTER THE NUMBER TO COUNT:"))
        print("THE COUNT IS :",num.count(value))
    elif choice==10:
        print("EXITING.....")
        print("THE FINAL LIST:",num)
        break
    else:
        print("INVALID CHOICE")
```
